**MATPLOTLIB PLOTTING**

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,8])
ypoints=np.array([3,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Draw a line')
plt.show()
```

```python
import matplotlib.pyplot as plt
X=range(1,50)
Y=[value*3 for value in X]
print("Values of X:")
print("Values in Y (thrice of X:)")
print(Y)
plt.plot(X,Y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw a line')
plt.show()
```

*DEFAULT X-POINTS*

```python
ypoint=np.array([3,8,1,10,5,7])
plt.plot(ypoint)
plt.show()
```

```python
#matplotlib markers

ypoint=np.array([3,8,1,10])
plt.plot(ypoint,marker='o')
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,marker='*')
plt.show()
```

*STRING FORMAT*

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,'o:r')
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,'o-r')
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,'o-.r')
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,'o:m')           #m for magenta color
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,linestyle='dotted')
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,linestyle='dashed')
plt.show()
```

*LINE COLOR*

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,color='r')
plt.show()
```

*LINE WIDTH*

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,linewidth='20.5')
plt.show()
```

**MULTIPLE LINES**

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```
